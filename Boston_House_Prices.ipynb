```python
#Import potrzebnych bibliotek
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#Pobranie danych do Dateframu
df_boston = pd.read_csv('https://raw.githubusercontent.com/scikit-learn/scikit-learn/main/sklearn/datasets/data/boston_house_prices.csv', header=1)
```

```python
df_boston
```

```python
#Zbudowanie funkcji w celu wyliczenia wyniku dla poszczególnych par
def oblicz(para1,para2):
    X = df_boston[[para1]]
    y = df_boston[para2]
    model = LinearRegression()
    model.fit(X, y)
    # obliczenie współczynnika korelacji
    corr_coef = df_boston[para1].corr(df_boston[para2])
    print("wykres zależności")
    plt.scatter(X, y)
    plt.plot(X, model.predict(X), color='red')
    plt.title(fr'{para1} kontra {para2} (corr={corr_coef:.2f})')
    plt.xlabel(para1)
    plt.ylabel(para2)
    plt.show()

    print(f'Współczynnik korelacji: {corr_coef:.2f}')
    
```

```python
pary = [("CRIM", "ZN"), ("CRIM", "INDUS"), ("CRIM", "RM"), ("AGE", "RM"), ("CRIM", "MEDV"), ("RAD", "MEDV"), ("AGE", "CRIM"), ("TAX", "AGE"), ("PTRATIO", "AGE"), ("AGE", "DIS")]
```

```python
for wynik in pary:
    oblicz(wynik[0],wynik[1])
```
